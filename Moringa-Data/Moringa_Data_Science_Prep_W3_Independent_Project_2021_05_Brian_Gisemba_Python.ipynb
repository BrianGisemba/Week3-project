```python
#Merging the 3 Telkom Datasets
df1=pd.read_csv('Telcom_dataset.csv')
df2=pd.read_csv('Telcom_dataset2.csv')
df3=pd.read_csv('Telcom_dataset3.csv')
frames=[df1,df2,df3]
result=pd.concat(frames)
result.head()
```

```python
#drop columns if the only contain missing values
result.dropna(axis=1, how='all')

```

```python
# Drop rows that contain less than five observations

result.dropna(thresh=5)
```

```python
#fill the missing values with 0
result.fillna(0)
```
